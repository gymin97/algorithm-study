```python
def calc_average (m,f):
    global score
    score = (m+f)/2
    return score

def clac_grade (score):
    global grade
    if score >= 90:
        grade = 'A'
    elif 80 <= score < 90:
        grade = 'B'
    elif 70 <= score < 80:
        grade = 'C'
    elif 60 <= score < 70:
        grade = 'D'
    else :
        grade = 'F'
    return grade
```
